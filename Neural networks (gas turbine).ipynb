# EDA

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

std_data = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
std_data
```

```python
from sklearn.model_selection import train_test_split
```

splitting the dataset into predictor and target

```python
x = std_data.drop('TEY',axis=1)
y = std_data[['TEY']]
```

```python
# Splitting data into test data and train data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=3)
```

# building the model

```python
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Build the model
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=x_train.shape[1]))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=1))
```

```python
# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=100, batch_size=32, verbose=1)
```

```python
# Evaluate the model on the test set
test_loss = model.evaluate(x_test, y_test, verbose=0)
print("Test loss:", test_loss)
```

```python
# Make predictions on new data
new_data = np.array([[6.8594,1007.9,96.799,3.5000,19.663,1059.2,550.00,10.605,3.1547,82.722]])
new_data = pd.DataFrame(scaler.fit_transform(new_data))
predictions = model.predict(new_data)
print("Predictions:", predictions)
```

# Hyper parameter tunning for batch_size and no. of epochs

```python
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import GridSearchCV

def build_model(optimizer='adam', activation='relu', neurons=64):
    model = Sequential()
    model.add(Dense(neurons, activation=activation, input_shape=(10,)))
    model.add(Dense(1))
    model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])
    return model

model = KerasRegressor(build_fn=build_model, verbose=0)

param_grid = {'optimizer': ['SGD', 'Adam', 'RMSprop'],
              'activation': ['relu', 'tanh', 'sigmoid'],
              'neurons': [16, 32, 64, 128, 256]}

grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1)
grid_result = grid.fit(x_train, y_train)

# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
